```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('dark_background')
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y=2*x
```

```python
plt.plot(x,y)
#plt.show() - this will eliminate the matplotlib object infor being returned

plt.title('String Title')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0,6)
plt.ylim(0,16)

plt.savefig('sampleplot.png')
```
