```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use('dark_background')
```

```python
x = np.linspace(0,11,10)
```

Changing the colour

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#65c79b',label='X vs X') #RGB HEX code

ax.plot(x,x+1,color='#81bccc',label= 'X vs X+1')

ax.legend()
```

Linewidth


```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#65c79b',lw=10) #linewidth
```

Linestyles

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#65c79b',linewidth=4,linestyle= '-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x,color='#65c79b',linewidth=4,linestyle= '-.')

#here we are feeding our custom dash sequence to be used in the plot
lines[0].set_dashes([5,2,5,2])
```

Markers 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#using markers we get to highlight each individual datapoint in our graph
#There are many marker styles available which are determined by the character fed to the marker parameter
ax.plot(x,x,color='#65c79b',linewidth=1,marker='o',ls='--',ms=10)
ax.plot(x,x+1,color='#42c72b',linewidth=1,marker='+',ls='--',ms=10)

#there are parameters to modify the marker to, go through that(ms - marker size)

plt.savefig('markers.png',bbox_inches='tight')
```
