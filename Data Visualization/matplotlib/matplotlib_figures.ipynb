```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('dark_background')
import numpy as np
```

```python
#Data
a = np.linspace(0,10,11)
```

```python
b= a**4
```

```python
x = np.arange(0,10)
y= 2*x
```

```python
#All these should be going into same cell
fig = plt.figure()

#Large Axes
axes1 = fig.add_axes([0,0,1,1])
#first two (0,0) we are specifying the bottom left corner. In this case we are starting it at the left bottom
#In case this was (0.5,0.5) we would have the axes placed somewhere in the middle of the axes

#The last two parameters are width and height. These are relative with respect to the height of the axes
#[1,1] means same size as the underlying axes. (0.5,0.5) would mean that the width would be half and height would be half the original axis


axes1.plot(a,b)

axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_xlim(0,12)
axes1.set_ylim(0,10000)
axes1.set_title('axis1')

#Small Axes

axes2= fig.add_axes([0.1,0.5,0.4,0.4])
axes2.plot(a,b)

axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_title('Zoomed in')

plt.show()
```

```python
fig = plt.figure(figsize=(2,2),dpi=200)
#figsize = (width,height)
#dpi - clarity/quality of figure

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('new_fig.png',bbox_inches = 'tight' )
# bbox_inches parameter is used so that the x and y-axis are saved too
```
