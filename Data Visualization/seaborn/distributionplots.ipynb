```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='darkgrid')
```

1.Rug plot 

One tick per data point on y axis

Y-axis is meaningless.

Helps identify outliers

2.KDE(kernal density estimation)

Estimating a continous distribution for a given datapoint


```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
#Rugplot
plt.figure(figsize=(2,2),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)

plt.savefig('rugplot.png')
```

```python
#Do not use distplot

sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=10,hue='level of education')
plt.savefig('histogram.png')
```

```python

sns.displot(data=df,x='salary',bins=10,color='blue',edgecolor='white')
```

```python
sns.histplot(data=df,x='salary')
```

```python
sns.displot(data=df,x='salary',kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)

sample_ages= np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages= pd.DataFrame(sample_ages,columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
sns.displot(data=sample_ages,x='age',bins=20,rug=True,kde=True)
plt.savefig('hist2.png')
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=0.6,shade=True)
#Clipping off values that aren't present in the distribution
#lower the value of bw_adjust, more the variance that is picked up. A high value would mean a high degress of underfitting
#shade fills up the area under the curve
```
