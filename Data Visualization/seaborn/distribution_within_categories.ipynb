```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

Violin plots - It is  the kde plot attached with its mirror image. The dot in the centre is its median.

Swarmplot and boxenplot are very rare. use them only if there is a requirement for those.

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
#Boxplot
sns.boxplot(data=df,y='math score',x='test preparation course')
```

```python
sns.boxplot(data=df,y='reading score',x='test preparation course')
```

```python
sns.boxplot(data=df,y='reading score',x='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1,1))
```

```python
#Violin plot

sns.violinplot(data=df,x='writing score',y='race/ethnicity',hue='test preparation course',split=True,inner='quartile')
#By having split=True one half of violin plot is one category other half is another
#Inner = quartile draws lines at each quartile

plt.legend(bbox_to_anchor=(1,1))
```

```python
#Violin plot

sns.violinplot(data=df,x='writing score',y='race/ethnicity',inner='stick')
#inner=stick will draw a line for every data point
```

```python
#Swarm plot
sns.swarmplot(data=df,x='math score',y='gender',hue='race/ethnicity',size=2,dodge=True)

#Hue paramater will be difficult to interpret since we are considering individual data points
#hence we go for dodge=true to split the plot based on the hue parameter

plt.legend(bbox_to_anchor=(1,1))
```

```python
#Boxen plot: Said to be more informative than normal box plot

sns.boxenplot(data=df,x='test preparation course',y='math score')
```
