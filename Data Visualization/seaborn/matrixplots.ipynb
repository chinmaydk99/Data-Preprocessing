```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style='darkgrid')
```

```python
#Heatmaps : Ideally all the values should be in the same unit 
#Helps identify regions/people with similar features.

#Seaborn can innately group together similar groups 

#Matrix plots

df=pd.read_csv("country_table.csv")

```

```python
df.head()
```

```python
df=df.set_index("Countries")
```

```python
df.head()
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,cmap='viridis')
```

```python
sns.clustermap(df.drop('Life expectancy',axis=1),linewidth=0.5,annot=True,cmap='viridis',col_cluster=False)
plt.savefig('clustermap.jpg')
```
