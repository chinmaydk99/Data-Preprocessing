```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='darkgrid')
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.jointplot(x='math score',y="reading score",data=df,kind='hex')
#kind parameter can be used to switch between hex, scatter of other plots
#jointplot plots the relation(scatter/hex plot) between the two categories and also the underlying distributions within these categories
```

```python
sns.jointplot(x='math score',y="reading score",data=df)
```

```python
sns.jointplot(x='math score',y="reading score",data=df,kind='kde',hue='gender')
```

```python
#Pairplots : it is CPU intensive
#Plots out histograms as well as scatter plots for every possible feature combination
#with pairplots, it is advisable to feed in only the columns we are interested in
sns.pairplot(data=df,hue='race/ethnicity')
```

```python
sns.pairplot(data=df,hue='race/ethnicity',diag_kind='hist',corner=True)

#corner=True eliminates the duplicates
#diag_kind parameter is used to ensure that the diagonal elements are histograms only
```
