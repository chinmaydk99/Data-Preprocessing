```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set(style='darkgrid')
```

```python
df=pd.read_csv("StudentsPerformance.csv")
```

Grids- Analogous to matplotlib subplots



```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch',row='test preparation course')
```

```python
grid = sns.PairGrid(df,hue='gender')
#When we use pairplot, internally it calls the pair grid functions
#The following functions are used to customise the plots along the diagonal, above the diagonal and below the diagonal

grid=grid.map_upper(sns.scatterplot)

grid= grid.map_lower(sns.kdeplot)

grid = grid.map_diag(sns.histplot)

grid=grid.add_legend()
```
