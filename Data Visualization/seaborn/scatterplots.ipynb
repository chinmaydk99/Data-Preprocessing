```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(5,5),dpi=200)
#coloring based on other column (continuous column - gradient , categorical - legend)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2')

#Palette parameter is used to set colormaps
```

```python
plt.figure(figsize=(5,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,size='salary',palette='Dark2',alpha=0.5)

#size paramater similar to color. Alpha is for transparency(when datapoints are stacked on top of each other)
```

```python
plt.figure(figsize=(5,5),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',alpha=1)

#style is particularly important in black and white plots

plt.savefig('scatterplot.png')
```
